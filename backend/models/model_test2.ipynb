```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
initial_data = pd.read_json('./playerStats.json')
initial_data
```

```python
initial_data = initial_data.drop(['player', 'team', 'game', 'pos', 'min', 'pFouls', 'steals', 'turnovers', 'blocks', 'plusMinus', 'comment'], axis = 1)
initial_data.dropna(inplace=True)
initial_data.reset_index(drop=True, inplace=True)
```

```python
last_3_games_avg = initial_data.groupby(np.arange(len(initial_data.index)) // 3).agg('mean')
last_3_games_avg = last_3_games_avg.add_prefix("last_3_avg_")
```

```python
new_df = initial_data.iloc[3::3]
new_df.reset_index(drop=True, inplace=True)
new_df
```

```python
if new_df.shape[0] > last_3_games_avg.shape[0]:
    new_df = new_df[:-1]
else:
    last_3_games_avg = last_3_games_avg[:-1]
```

```python
new_df = new_df[['points', 'totReb', 'assists']]
pd.concat([last_3_games_avg, new_df], axis = 1)
```
